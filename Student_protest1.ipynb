```python
import pandas as pd
from nrclex import NRCLex


# Function to extract sentiment and emotions separately
def extract_sentiments_emotions(text):
    text_obj = NRCLex(text)  # Analyze text

    # Extract emotion scores
    emotions = text_obj.raw_emotion_scores

    # Calculate sentiment manually
    positive_score = sum([emotions.get(em, 0) for em in ['joy', 'trust']])  # NRC defines joy & trust as positive
    negative_score = sum([emotions.get(em, 0) for em in ['anger', 'disgust', 'fear', 'sadness']])  # Negative emotions

    # Determine sentiment label
    if positive_score > negative_score:
        sentiment_label = "Positive"
    elif negative_score > positive_score:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"

    # Convert emotions into a readable format
    emotion_label = ', '.join(emotions.keys()) if emotions else "Neutral"

    return sentiment_label, emotion_label

# Apply function to extract sentiment and emotions
df[['Sentiment', 'Emotions']] = df['Comment_processed'].apply(lambda x: pd.Series(extract_sentiments_emotions(x)))

```

```python
import pandas as pd
from collections import Counter
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Define emotion columns
emotion_cols = ['joy', 'positive', 'anger', 'disgust', 'fear',
                'negative', 'trust', 'anticipation', 'sadness', 'surprise']

# Function to extract top terms for a given emotion
def get_top_terms(df, emotion, text_column='Comment_processed', top_n=10):
    relevant_texts = df[df[emotion] == 1][text_column].dropna().tolist()
    all_words = []
    for text in relevant_texts:
        tokens = word_tokenize(text.lower())
        all_words.extend(tokens)
    most_common = Counter(all_words).most_common(top_n)
    return ', '.join([word for word, count in most_common])

# Prepare the table
results = {
    'Emotion': [],
    'Top Terms': [],
    'Mean': [],
    'Standard Deviation': []
}

for emo in emotion_cols:
    results['Emotion'].append(emo)
    results['Top Terms'].append(get_top_terms(df, emo))
    results['Mean'].append(df[emo].mean())
    results['Standard Deviation'].append(df[emo].std())

# Create and display the DataFrame
emotion_summary_df = pd.DataFrame(results)

emotion_summary_df
```
